```python
import xml.dom.minidom
```

```python
doc = xml.dom.minidom.parse("7596.male.26.Internet.Scorpio.xml");
```

```python
post = doc.getElementsByTagName("post")
n=post.length
n
```

```python
li=[]
for i in range(0,n):
    memoryElem = doc.getElementsByTagName('post')[i]
    li.append( [node.data for node in memoryElem.childNodes] )
#print(li)
li
```

```python

for i in range(0,len(li)):
    for j in range(0,len(li[i])):
        li[i][j] = li[i][j].strip()
    
li
```

```python
from nltk.tokenize import word_tokenize
tokens=[]
for i in range(0,len(li)):
    for text in li[i]:      
        token= word_tokenize(text)
        tokens.append(token)
tokens
    
```

```python
words=[]
count=[]
for i in range(0,len(tokens)):
    uniqueWords = sorted(set(tokens[i]))
    for word in uniqueWords:
        words.append(word)
        count.append(tokens[i].count(word))
        
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(words,columns =['Word']) 
df2= pd.DataFrame(count,columns =['Frequency']) 
```

```python
df.tail()
```

```python
df2.tail()
```

```python
merged = pd.concat([df,df2], axis=1,ignore_index=True)
```
